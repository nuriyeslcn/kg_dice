```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Initial', 'Phase I', 'Phase II', 'Phase III', 'DiCE Genes\nGroup 1', 'DiCE Genes\nGroup 2']
counts = [30715, 14044, 5608, 4430, 1332, 1453]

colors = ['#4a7fa5', '#5a9ab5', '#6ab8b0', '#7dd4a0', '#e07b5a', '#e09a5a']

fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(labels, counts, color=colors, width=0.5, edgecolor='white', linewidth=0.5)

for bar, count in zip(bars, counts):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 300,
            f'{count:,}', ha='center', va='bottom', fontsize=11, fontweight='bold', color='#333333')

ax.axvline(x=3.5, color='#e07b5a', linestyle='--', linewidth=1.5, alpha=0.7)
ax.text(3.6, max(counts) * 0.9, 'DiCE Genes', color='#e07b5a', fontsize=9)

ax.set_ylabel('Number of Genes', fontsize=11, color='#555555')
ax.set_title('DiCE Gene Selection', fontsize=14, fontweight='bold', pad=15)
ax.set_ylim(0, max(counts) * 1.15)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(axis='x', labelsize=10)
ax.tick_params(axis='y', labelsize=9)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))
info_text = "Group 1: distance & correlation as weights\nGroup 2: only distance as weight"

ax.text(0.98, 0.95, info_text,
        transform=ax.transAxes,
        fontsize=8.5,
        verticalalignment='top',
        horizontalalignment='right',
        bbox=dict(boxstyle='round,pad=0.5', facecolor='#f5f0eb', edgecolor='#e07b5a', linewidth=1.2),
        color='#333333',
        linespacing=1.6)

plt.tight_layout()
plt.savefig('dice_pipeline.png', dpi=300, bbox_inches='tight')
plt.show()
```
