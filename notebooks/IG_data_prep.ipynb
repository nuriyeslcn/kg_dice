## Phase II Preparation:
+ This notebook is where we prepare the raw normalized expression data from [GSE252168 Study](https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE252168) to be used for Information Gain (IG) calculation in Phase II. In the part 2 of this notebooks, we also prepare a dataframe where patient IDs are mapped to their corresponding health status (Control vs Lung Cancer) to later be used for IG calculation as well.

## Part 1: Prepare the normalized-raw expression matrix to Information Gain (Phase II)
+ Normalized-raw expression dataset of our study stored as [GSE252168_Matrix_normalized](../data/GSE252168/GSE252168_Matrix_normalized.txt) in the data folder.
+ Filtered DEA results from Phase I is stored as [dea_filtered](../data/GSE252168/dea_filtered.tsv) in the data folder. 

```python
import pandas as pd 
```

```python
# Get IDs from DEA filtered dataset to use filter my raw expression data
dea_filt = pd.read_csv("../data/GSE252168/dea_filtered.tsv", sep="\t", usecols=[0])
print(dea_filt.shape)
dea_filt.head()
```

```python
# Get only IDs
id_dea = dea_filt["ID"].to_list()
print(len(id_dea))
print(id_dea)
```

```python
# Upload raw & Normalized expression data from GEO
raw_norm = pd.read_csv("../data/GSE252168/GSE252168_Matrix_normalized.txt", sep="\t")
print(raw_norm.shape)
raw_norm.head()
```

```python
# Filtered normalized raw data via genes coming from DEA
raw_filt = raw_norm[raw_norm["ID_REF"].isin(id_dea)].copy()
print(raw_filt.shape)
raw_filt.head()
```

```python
# Get column names not containing Detection phrase since we only need gene expression values
new_cols = raw_filt.columns[~raw_filt.columns.str.contains("Detection")]

# Filter normalized row dataset for only patient (control & cancer) expression columns
row_filt2 = raw_filt[new_cols].copy()
print(row_filt2.shape)
row_filt2.head()
```

```python
## Transpose the dataframe to get genes as columns and patients as rows
# row_filt2.transpose().to_csv('../data/GSE252168/row_norm_filt_v1.tsv', sep='\t')
```

```python
# Re-upload the raw, normalized and filtered dataset to set column names correctly after transposition
row_final = pd.read_csv("../data/GSE252168/row_norm_filt_v1.tsv", sep="\t", header=1)
print(row_final.shape)
row_final.tail()
```

```python
# Replace "401_" with "NLCB_" to match other mappings
row_final["ID_REF"] = row_final["ID_REF"].apply(lambda x: x.replace("401_", "NLCB_"))
```

```python
# Verify the replace in the dataframe
row_final.tail()
```

```python
# Finally order the dataframe to have same order with ids to patients status mapping later
row_final.sort_values("ID_REF", inplace = True)
```

```python
print(row_final.shape)
row_final.tail()
```

## Part 2: Map patient IDs to health conditions
- Metadata of our study can be found in the data folder as [metadata_table](../data/GSE252168/metadata_table.html). 
- We will use this metadata to create a mapping between patient IDs and their corresponding health conditions (Control vs Lung Cancer) for later use in IG calculation.
- Mapping itself is stored as [id_to_health_status](../data/GSE252168/id_to_health_status.tsv) in the data folder.

```python
# Check the metadata before filtering for patient IDs and health status
metadata = pd.read_html("../data/GSE252168/metadata_table.html")
metadata1 = metadata[0]
print(metadata1.shape)
metadata1.head()
```

```python
# Get corresponding health condition for each patient in a different dataframe
id_to_cond = metadata1.filter(["Title", "Status"], axis="columns")
print(id_to_cond.shape)
id_to_cond.tail()
```

```python
# Remove the "blood" phrase since our expression map IDs do not contain that phrase
id_to_cond["Title"] = id_to_cond["Title"].apply(lambda x: x.replace("blood_", ""))
id_to_cond.head()
```

```python
# Replace "case" status via "lung cancer"
id_to_cond.replace("case","lung cancer",inplace=True)
id_to_cond.tail()
```

```python
# Check how many controls and lung cancer patients there to verify 
id_to_cond["Status"].value_counts()
```

```python
# Sort the ID column to get same order with expression matrix
id_to_cond.sort_values("Title", inplace= True)
```

```python
# Get id order from the expression matrix
patient_ids = pd.read_csv("../data/GSE252168/row_filtered.tsv", sep="\t", usecols= ["ID_REF"])
```

```python
# Check whether the order of the samples match 
match_order = [i for i, j in zip(id_to_cond["Title"], patient_ids["ID_REF"]) if i == j]
len(match_order)
```
