# Phase III Part 1: modification of PPI network with gene expression correlations

```python
import pandas as pd
```

## Part 1: Filter Gene Expression Data 
+ First, filter the gene expression data to get only the genes that we interested in, which are the genes we got after IG analysis.
+ Then, get separate the control and patient dataframes for further analysis.

```python
# Get the gene IDs in a list to filter gene expression matrix
gene_ids=pd.read_csv("../data/GSE252168/IG_results.tsv", sep="\t", usecols=[0])
gene_ids=gene_ids["Unnamed: 0"].to_list()
print(len(gene_ids))
```

```python
# Upload the expression matrix
exp_df = pd.read_csv("../data/GSE252168/row_filtered.tsv",sep="\t")
print(exp_df.shape)
exp_df.head()
```

```python
# Filter the expression matrix with our filtered gene pool
exp_df = exp_df.loc[:, exp_df.columns.isin(gene_ids) | (exp_df.columns == exp_df.columns[0])]
print(exp_df.shape)
exp_df.head()
```

```python
# Upload the patient ID mapping file
id_to_cond =pd.read_csv("../data/GSE252168/id_to_health_status.tsv", sep="\t")
id_to_cond.head()
```

```python
control_ids = list(id_to_cond[id_to_cond["Status"] == "control"]["Title"])
print(f"Number of control: {len(control_ids)}")
patient_ids = list(id_to_cond[id_to_cond["Status"] == "lung cancer"]["Title"])
print(f"Number of lung cancer patient: {len(patient_ids)}")
```

```python
# Get controls via filtering for their IDs
control_df = exp_df[exp_df['ID_REF'].isin(control_ids)]
print(control_df.shape)
control_df.head()
```

```python
# Get patients via filtering for their IDs
patient_df = exp_df[exp_df['ID_REF'].isin(patient_ids)]
print(patient_df.shape)
patient_df.head()
```

## Part 2: Pearson Correlation Analysis & Distance Matrix

```python
# Pearson correlation
controls_corr = control_df.corr(method='pearson', numeric_only=True)
patients_corr = patient_df.corr(method='pearson', numeric_only=True)
```

```python
# Convert the correlation matrices to distance matrices
controls_dist = 1 - abs(controls_corr)
patients_dist = 1 - abs(patients_corr)
```

```python
controls_dist
```

```python
# Upload the DEA result table to get gene symbols for corresponding gene IDs
dea_df = pd.read_csv("../data/GSE252168/dea_filtered.tsv", sep="\t", usecols=[0,1])
dea_df.head()
```

```python
# Filtered gene pool
filtered_genes = pd.read_csv("../data/GSE252168/IG_results.tsv", sep="\t", usecols=[0])
filtered_genes = filtered_genes["Unnamed: 0"].to_list()
print(len(filtered_genes))
```

```python
# Filter the DEA result table to get the gene symbols for our filtered gene pool
dea_df = dea_df[dea_df["ID"].isin(filtered_genes)]
dea_df.shape
```

```python
# Create a dict where we have mapping from gene id to gene symbol
gene_id_to_symbol = dict(zip(dea_df["ID"], dea_df["Gene.symbol"]))
print(gene_id_to_symbol)
```

```python
# Map gene IDs to gene symbols in the distance matrices
controls_dist = controls_dist.rename(index=gene_id_to_symbol, columns=gene_id_to_symbol)
patients_dist = patients_dist.rename(index=gene_id_to_symbol, columns=gene_id_to_symbol)
```
