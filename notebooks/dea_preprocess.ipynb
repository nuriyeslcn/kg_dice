## Phase I: construction of gene candidate pool by DEA
+ This notebooks is where we filtered DEA results which comes from Phase I. We used [GEO2R](https://www.ncbi.nlm.nih.gov/geo/geo2r/) for DEA of our [GSE252168 Study](https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE252168). Raw DEA dataset coming from GEO2R is stored as [GSE252168.top.table](../data/GSE252168/GSE252168.top.table.tsv) in the data folder. 

```python
import pandas as pd
```

```python
# Number of samples: 303 (123 lung cancer patients, 180 controls)
dea_results = pd.read_csv("../data/GSE252168.top.table.tsv", sep="\t")
print(dea_results.shape)
dea_results.head()
```

```python
# Remove entries where we are missing Gene Symbol since we cannot use them later via PPI network
dea_df1 = dea_results[dea_results["Gene.symbol"].notna()].copy()
# Also apply threshold '0.05' for adjusted p values
dea_df2 = dea_df1[dea_df1["adj.P.Val"] < 0.05].copy()
print(dea_df2.shape)
dea_df2.head()
```
